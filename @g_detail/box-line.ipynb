```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy import stats
```

```python
model = 'TcnED'
data = 'ASD'
func = 'ICLR21'
```

```python
from matplotlib import pyplot
plt.style.use('seaborn-whitegrid')
palette = pyplot.get_cmap('Set1')
```

```python
def plot_single(pindex, dtindex, dfindex, title):
    truelist = []
    falselist = []
    true = df1[df1.label == 0][pindex].values[dtindex]
    false = df1[df1.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df2[df2.label == 0][pindex].values[dtindex]
    false = df2[df2.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df3[df3.label == 0][pindex].values[dtindex]
    false = df3[df3.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df4[df4.label == 0][pindex].values[dtindex]
    false = df4[df4.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df5[df5.label == 0][pindex].values[dtindex]
    false = df5[df5.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df6[df6.label == 0][pindex].values[dtindex]
    false = df6[df6.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df7[df7.label == 0][pindex].values[dtindex]
    false = df7[df7.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df8[df8.label == 0][pindex].values[dtindex]
    false = df8[df8.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    true = df9[df9.label == 0][pindex].values[dtindex]
    false = df9[df9.label == 1][pindex].values[dfindex]
    true = np.mean(true, axis=1)
    false = np.mean(false, axis=1)
    truelist.append(true)
    falselist.append(false)

    truelist = np.array(truelist)
    falselist = np.array(falselist)
    iters = [0,1,2,3,4,5,6,7,8]

    color=palette(0)
    avg=np.mean(truelist,axis=1)
    std=np.std(truelist,axis=1)
    r1 = list(map(lambda x: x[0]-x[1], zip(avg, std)))#上方差
    r2 = list(map(lambda x: x[0]+x[1], zip(avg, std)))#下方差
    plt.plot(iters, avg, label="Clean", color=color, linewidth=3.0)
    plt.fill_between(iters, r1, r2, color=color, alpha=0.2)

    color=palette(1)
    avg=np.mean(falselist,axis=1)
    std=np.std(falselist,axis=1)
    r1 = list(map(lambda x: x[0]-x[1], zip(avg, std)))#上方差
    r2 = list(map(lambda x: x[0]+x[1], zip(avg, std)))#下方差
    plt.plot(iters, avg, label="Polluted",color=color, linewidth=3.0)
    plt.fill_between(iters, r1, r2, color=color, alpha=0.2)

    plt.legend()
    plt.xlabel('Epoch',fontsize=22)
    plt.ylabel('Parameter Importance',fontsize=22)
    plt.title(str(title))

    plt.show()
```

```python
dfindex = [0]
for i in range(10):
    pindex = ['5','6']
    dfindex = np.array([i])
    plot_single(pindex, dtindex, dfindex, dfindex)
```
